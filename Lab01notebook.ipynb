```python
import numpy as np
import matplotlib.pyplot as plt
# Parameters
N = 20 # number of bits
sps = 10 # samples per symbol
fc = 2 # carrier frequency (Hz)
fs = sps * 10 # sampling frequency
t = np.arange(0, N, 1/sps) # time vector
# 1. Source bits
bits = np.random.randint(0, 2, N)
symbols = 2*bits - 1 # BPSK mapping: 0-1, 1+1
baseband = np.repeat(symbols, sps)
# 2. Carrier
carrier = np.cos(2*np.pi*fc*t)
# 3. Modulated signal
tx = baseband * carrier
# 4. Add simple noise
noise = 0.3*np.random.randn(len(tx))
rx = tx + noise
# 5. Demodulate (multiply by carrier again)
demod = rx * carrier
# ---- PLOTS ----
plt.figure(figsize=(10,6))
plt.subplot(4,1,1); plt.plot(baseband); plt.title("Digital the baseband")
plt.subplot(4,1,2); plt.plot(carrier); plt.title("Analog, the carrier ")
plt.subplot(4,1,3); plt.plot(tx); plt.title("transmitted signal")
plt.subplot(4,1,4); plt.plot(demod); plt.title("Analog with noise, received after demodulation.")
plt.tight_layout()
plt.show()
```
